```python
from optstoicpy.core.drawpathway import draw_pathway
from optstoicpy.core.pathway import Pathway
import pandas as pd
from IPython.display import display, Image
```

## 1. Draw a pathway frm an Excel input file `reaction_list.xlsx`

```python
# Read the "test" worksheet from the excel file "reaction_list.xlsx"
df = pd.read_excel("reaction_list.xlsx", sheet_name="test")

# This create a "Pathway" instance
test_pathway = Pathway(name='EMP',
                    reaction_ids=df.reaction_id.tolist(),
                    fluxes=df.flux.tolist())

# Create a png image
figure_filename = 'test_pathway'
figure_ext = "png"

g = draw_pathway(
    test_pathway, 
    imageFileName=figure_filename, 
    imageFormat=figure_ext, 
    graphTitle=test_pathway.name,
    cleanup=True, 
    darkBackgroundMode=False,
    width=5,
    height=5)

Image(figure_filename + "." + figure_ext)
```
